```python
import json 
with open("IN_category_id.json") as f:
    data = json.load(f)

category_mapping = {}
for item in data["items"]:
    category_mapping[int(item["id"])] = item["snippet"]["title"]
```

```python
df['category_name'] = df['category_id'].map(category_mapping)
```

```python
df.head()
```

```python
top_category = df["category_name"].value_counts().sort_values(ascending=False)
```

```python
top_category
```

```python
category_views = df.groupby('category_name')['views'].sum().sort_values(ascending=False)
```

```python
category_views
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
category_views.plot(kind='bar', color='orange')
plt.title("Total Views per Category")
plt.xlabel("Category")
plt.ylabel("Total Views")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.yscale('log')

plt.show()
```

```python
correlation = df["views"].corr(df["likes"])
print("Correlation between views and likes:", correlation) 
```

```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

all_tags = df['tags'].dropna().apply(lambda x: x.split('|')) 
flat_tags = [tag.strip() for sublist in all_tags for tag in sublist]
tag_counts = Counter(flat_tags)
most_common_tags = tag_counts.most_common(20)
print("Top 20 tags:", most_common_tags)
wordcloud = WordCloud(width=800, height=400,background_color='white').generate_from_frequencies(tag_counts)

# Plot
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
